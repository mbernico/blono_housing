# What's an EDA?
EDA or exploritory data analysis is a process where we examine our data.  We *think* something in our dataset might be valuable.  Now we need to look through it to learn more about what might be helpful.

Things I personally look at:
*  Overall shape (# of variables and observations)
*  How much data is missing?
*  Do I need to clean or filter the data?
*  Variable types (continuous, categorical, ordinal)

Lets do some EDA!

### So, which of these variables are going to most likely predict price?

Pearson's Correlation Coefficient is a handy formula for calculating how much two variables move together.  [Here's a link](https://en.wikipedia.org/wiki/Pearson_correlation_coefficient#Definition)

What does this mean, without all the fancy math?

Lets say we had two variables
X = 1, 2, 4
Y = 10, 20, 40
X and Y would have a Pearson Correlation Coefficient of 1.  They're perfectly correlated.

Now, lets try it again.
X = 1, 2, 4
Y = 40, 20, 10

Now X and Y would be inversly correlated.  The PCC would be -1

* So, what would a PCC of .5 mean?
* How about .75

```python
corr = sfh[["bedrooms", "baths_full", "baths_half", "garage","sqft","price"]].corr()
print(corr)
```

```python
cmap = cmap=sns.diverging_palette(5, 250, as_cmap=True)

corr.style.background_gradient(cmap, axis=1)\
    .set_properties(**{'max-width': '80px', 'font-size': '10pt'})\
    .set_precision(2)
```
